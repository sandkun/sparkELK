### 유튜브 크롤링 셀레니움

```python
req = requests.get("https://www.youtube.com/channel/UCDBAVzfX3yZ1hah0FHnOoaA")

html = req.text

soup = BeautifulSoup(html, "html.parser")

#구독자
subscriber = soup.select(".yt-subscription-button-subscriber-count-branded-horizontal")[0].text
#이름
name = soup.select("meta[content='괴물쥐 유튜브']")[0].get("content")
#채널 썸네일 이미지
thumbnail = soup.select(".appbar-nav-avatar")[0].get("src")

videos_html = requests.get("https://www.youtube.com/channel/UCDBAVzfX3yZ1hah0FHnOoaA/videos").text
videos = BeautifulSoup(videos_html, "html.parser")
sum = 0
for x in videos.select(".yt-lockup-content"):
    sum += int(x.select(".yt-lockup-meta-info li")[0].text[4:][:-1].replace(",",""))
print(sum, subscriber, name)
```

```python
delay=3
browser = Chrome("/Users/joyumin/MyProject/youtubeTrand/chromedriver")
browser.implicitly_wait(delay)
start_url = 'https://www.youtube.com/channel/UCDBAVzfX3yZ1hah0FHnOoaA/videos'
browser.get(start_url)
browser.maximize_window()

body = browser.find_element_by_tag_name('body')

num_of_pagedowns = 20

while num_of_pagedowns:
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(2)
    num_of_pagedowns -= 1

html0 = browser.page_source
html = BeautifulSoup(html0, 'html.parser')
```

##### youtuber
{ <br>

&nbsp; name : ~, <br>
&nbsp; timestamp : ~, <br>
&nbsp; subscriber : ~, <br>
&nbsp; videos : [ <br>
&nbsp;&nbsp;     title : ~, <br>

&nbsp; ] <br>
}

__값 넣기__

```python
doc = {
    "id" : "UCDBAVzfX3yZ1hah0FHnOoaA", 
    "name" : name,
    "subscriber" : subscriber,
    "date" : time.time(),
    "views" : views
}
es.index(index="youtuber", body=doc)
```

__값 찾기__

```python
es.search(
    index="youtuber", 
    body = {
        "query":{"match":{"id":"UCDBAVzfX3yZ1hah0FHnOoaA"}}
    }
)
```
