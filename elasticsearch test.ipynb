### 유튜브 크롤링 셀레니움

```python
req = requests.get("https://www.youtube.com/channel/UCEM9IuTAFtTHs3Ca27xaZiw")
print(req)
html = req.text

soup = BeautifulSoup(html, "html.parser")
# print(soup)
#구독자
subscriber = soup.select(".subscribed")
if subscriber != []:
    subscriber = subscriber[0].text
else:
    subscriber = -1
print(subscriber)
#이름
name = soup.select("meta[itemprop='name']")[0].get('content')
print(name)
#채널 썸네일 이미지
thumbnail = soup.select(".appbar-nav-avatar")[0].get("src")

videos_html = requests.get("https://www.youtube.com/channel/UCDBAVzfX3yZ1hah0FHnOoaA/videos").text
videos = BeautifulSoup(videos_html, "html.parser")
sum = 0
for x in videos.select(".yt-lockup-content"):
    sum += int(x.select(".yt-lockup-meta-info li")[0].text[4:][:-1].replace(",",""))
print(sum, subscriber, name)
```

```python
delay=3
options = webdriver.ChromeOptions()
options.add_argument("headless")
options.add_argument("disable-gpu")
options.add_argument('window-size=1920x1080')

browser = Chrome("/Users/joyumin/MyProject/youtubeTrand/chromedriver", chrome_options=options)
browser.implicitly_wait(delay)
start_url = 'https://www.youtube.com/channel/UCDBAVzfX3yZ1hah0FHnOoaA/videos'
browser.get(start_url)
browser.maximize_window()

body = browser.find_element_by_tag_name('body')

last_scrollY = -1
while True:
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(2)
    scrollY = browser.execute_script('return window.scrollY')
    if last_scrollY == scrollY:
        break
    last_scrollY = scrollY
print("크롤링 완료")
html0 = browser.page_source
html = BeautifulSoup(html0, 'html.parser')
browser.close()
```
